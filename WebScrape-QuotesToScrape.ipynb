```python
from bs4 import BeautifulSoup #imports BeautifulSoup function
import requests #requests to get html data
import csv #writes to csv file

#scrape for quotes.toscrape.com
scrape_page = requests.get("https://quotes.toscrape.com/") 
soup = BeautifulSoup(scrape_page.text, "html.parser") 

#Finds attributes with text (quotes) and author
quotes = soup.findAll("span", attrs={"class":"text"}) 
authors = soup.findAll("small", attrs={"class":"author"}) 

#Opens new csv file then initialise's write
file = open("scraped_quotes.csv", "w")
writer = csv.writer(file)

#writer function
writer.writerow(["Quotes", "Authors"])

#Loop to loop over all quotes then close
for quote, author in zip(quotes, authors):
    #print(quote.text + " - " + author.text) - used to print quotes with authors if needed
    writer.writerow([quote.text, author.text])
file.close()
```
